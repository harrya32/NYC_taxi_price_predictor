```python
sample = df18.sample(10000000)
sample['cost'] = np.log(sample['cost'])
sample['approx_dist'] = np.log(sample['approx_dist'])
```

```python
columns = ['PU_crashes', 'DO_crashes', 'approx_dist', 'PU_rides_in_zone', 'DO_rides_in_zone', \
          'PUBorough', 'DOBorough', 'tod', 'month', 'dow', 'DailyAverageDryBulbTemperature', \
          'DailyAverageRelativeHumidity', 'DailyAverageWindSpeed', 'DailyPrecipitation', \
           'DailySnowDepth', 'DailySnowfall', 'cost']

continuous = ['PU_crashes', 'DO_crashes', 'approx_dist', 'PU_rides_in_zone', 'DO_rides_in_zone', \
             'DailyAverageDryBulbTemperature','DailyAverageRelativeHumidity', 'DailyAverageWindSpeed', \
             'DailySnowDepth', 'DailySnowfall']
scale = StandardScaler()
```

```python
sample = sample[columns]
```

```python
sample[continuous] = scale.fit_transform(sample[continuous])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
test_sample = df19.sample(1000, random_state = 1)
test_sample['cost'] = np.log(test_sample['cost'])
test_sample['approx_dist'] = np.log(test_sample['approx_dist'])
test_sample = test_sample[columns]
test_sample[continuous] = scale.transform(test_sample[continuous])
```

```python
#ADDITIVE
```

```python
model1 = ols('cost ~ PU_crashes + DO_crashes + approx_dist + PU_rides_in_zone + \
DO_rides_in_zone + DailyAverageDryBulbTemperature + DailyAverageRelativeHumidity + \
DailyAverageWindSpeed + DailySnowDepth + DailySnowfall + PUBorough + DOBorough + \
tod + dow + month', data = sample)
fit1 = model1.fit()
print(fit1.summary())
```

```python
pred = fit1.predict(test_sample)
```

```python
mean_absolute_error(np.exp(pred), np.exp(test_sample['cost']))
```

```python
mean_squared_error(np.exp(test_sample['cost']), np.exp(pred), squared = False)
```

```python
#ADDING INTERACTION
```

```python
model2 = ols('cost ~ PU_crashes + DO_crashes + approx_dist + PU_rides_in_zone + \
DO_rides_in_zone + DailyAverageDryBulbTemperature + DailyAverageRelativeHumidity + \
DailyAverageWindSpeed + DailySnowDepth + DailySnowfall + PUBorough + DOBorough + \
tod + dow + month + PUBorough*approx_dist + DOBorough*approx_dist + tod*dow + tod*PUBorough + \
tod*DOBorough + dow*PUBorough + dow*DOBorough', data = sample)
fit2 = model2.fit()
print(fit2.summary())
```

```python
pred = fit2.predict(test_sample)
```

```python
mean_absolute_error(np.exp(pred), np.exp(test_sample['cost']))
```

```python
mean_squared_error(np.exp(test_sample['cost']), np.exp(pred), squared = False)
```

```python
#REMOVING INSIGNIFICANT ATTRIBUTES AND ONES W VERY SMALL COEFFICIENTS
```

```python
#remove humidity
model3 = ols('cost ~ PU_crashes + DO_crashes + approx_dist + PU_rides_in_zone + \
DO_rides_in_zone + DailyAverageDryBulbTemperature + DailyAverageWindSpeed + \
DailySnowDepth + DailySnowfall + PUBorough + DOBorough + \
tod + dow + month + PUBorough*approx_dist + DOBorough*approx_dist + tod*dow + tod*PUBorough + \
tod*DOBorough + dow*PUBorough + dow*DOBorough', data = sample)
fit3 = model3.fit()
print(fit3.summary())
```

```python
pred = fit3.predict(test_sample)
```

```python
mean_absolute_error(np.exp(pred), np.exp(test_sample['cost']))
```

```python
mean_squared_error(np.exp(test_sample['cost']), np.exp(pred), squared = False)
```

```python
#get final model statistics
```

```python
MAE = []
RMSE = []
for i in range(100):
    x = df19.sample(10000)
    x['cost'] = np.log(x['cost'])
    x['approx_dist'] = np.log(x['approx_dist'])
    
    x[continuous] = scale.transform(x[continuous])
    
    p = fit3.predict(x)
    MAE.append(mean_absolute_error(np.exp(p), np.exp(x['cost'])))
    RMSE.append(mean_squared_error(np.exp(x['cost']), np.exp(p), squared = False))
```

```python
import folium
import geopandas as gpd
```

```python
geo_sample = df19.groupby('PULocationID').sample(n = 10, replace = True)
```

```python
geo_sample['approx_dist'] = np.log(geo_sample['approx_dist'])
geo_sample[continuous] = scale.transform(geo_sample[continuous])
```

```python
predictions = np.exp(fit3.predict(geo_sample))
```

```python
geo_sample['abs_err'] = abs(geo_sample['cost'] - predictions)
```

```python
sf = gpd.read_file('../preprocessed_data/zones/fixed_taxi_zones.shp')
```

```python
gdf = gpd.GeoDataFrame(pd.merge(geo_sample, sf, left_on='PULocationID', right_on='LocationID')).drop('PULocationID',axis=1)
```

```python
geoJSON = gdf[['LocationID','geometry']].drop_duplicates('LocationID').to_json()
```

```python
means = gdf.groupby('LocationID').mean().reset_index()
```

```python
m = folium.Map(location=[40.66, -73.94], tiles="Stamen Terrain", zoom_start=10)

# refer to the folium documentations on how to plot aggregated data.
m.add_child(folium.Choropleth(
    geo_data=geoJSON,
    data = means,
    columns = ["LocationID", "abs_err"],
    key_on="properties.LocationID",
    fill_color="YlOrBr",
    fill_opacity=0.7,
    line_opacity=0.2,
    name='choropleth',
    legend_name = "Absolute Prediction Error ($)"
))

#m.save('../plots/abs_pred_err.html')
m
```
